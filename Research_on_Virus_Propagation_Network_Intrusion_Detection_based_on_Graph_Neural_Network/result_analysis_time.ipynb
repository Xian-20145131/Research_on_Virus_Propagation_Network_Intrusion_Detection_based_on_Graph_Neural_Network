```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='whitegrid')
```

```python
result_nslkdd_binary = pd.read_csv('./result/nslkdd/binary/total_result_nslkdd.csv')
result_kdd99_binary = pd.read_csv('./result/kdd99/binary/total_result_kdd99.csv')
result_nslkdd_five = pd.read_csv('./result/nslkdd/five/total_result_nslkdd.csv')
result_kdd99_five = pd.read_csv('./result/kdd99/five/total_result_kdd99.csv')
```

```python
result_nslkdd_five.groupby(['feature_num']).mean()[['training_time', 'testing_time']]
```

```python
result_nslkdd_time_binary = result_nslkdd_binary.groupby(['feature_num']).mean()[['training_time', 'testing_time']]
result_kdd99_time_binary = result_kdd99_binary.groupby(['feature_num']).mean()[['training_time', 'testing_time']]
result_nslkdd_time_five = result_nslkdd_five.groupby(['feature_num']).mean()[['training_time', 'testing_time']]
result_kdd99_time_five = result_kdd99_five.groupby(['feature_num']).mean()[['training_time', 'testing_time']]
```

```python
fig = plt.figure(figsize=(16, 10))

ax1 = plt.subplot(2, 2, 1)
plt.plot(result_kdd99_time_binary.index, result_kdd99_time_binary['training_time'], label='KDD CUP 99 (BATCH_SIZE=256)')
plt.plot(result_nslkdd_time_binary.index, result_nslkdd_time_binary['training_time'], label='NSL-KDD (BATCH_SIZE=256)')
plt.legend(loc='upper left')
plt.xlabel('Feature number\n(a)')
plt.ylabel('Time (s)')
plt.title('Training time in binary task')

ax2 = plt.subplot(2, 2, 2)
plt.plot(result_kdd99_time_binary.index, result_kdd99_time_binary['testing_time'], label='KDD CUP 99 (BATCH_SIZE=256)')
plt.plot(result_nslkdd_time_binary.index, result_nslkdd_time_binary['testing_time'], label='NSL-KDD (BATCH_SIZE=256)')
plt.legend(loc='upper left')
plt.xlabel('Feature number\n(b)')
plt.ylabel('Time (s)')
plt.title('Testing time in binary task')

ax3 = plt.subplot(2, 2, 3)
plt.plot(result_kdd99_time_five.index, result_kdd99_time_five['training_time'], label='KDD CUP 99 (BATCH_SIZE=1024)')
plt.plot(result_nslkdd_time_five.index, result_nslkdd_time_five['training_time'], label='NSL-KDD (BATCH_SIZE=1024)')
plt.legend(loc='upper left')
plt.xlabel('Feature number\n(c)')
plt.ylabel('Time (s)')
plt.title('Training time in five-class task')

ax4 = plt.subplot(2, 2, 4)
plt.plot(result_kdd99_time_five.index, result_kdd99_time_five['testing_time'], label='KDD CUP 99 (BATCH_SIZE=1024)')
plt.plot(result_nslkdd_time_five.index, result_nslkdd_time_five['testing_time'], label='NSL-KDD (BATCH_SIZE=1024)')
plt.legend(loc='upper left')
plt.xlabel('Feature number\n(d)')
plt.ylabel('Time (s)')
plt.title('Testing time in five-class task')

plt.subplots_adjust(hspace=0.35)
```
