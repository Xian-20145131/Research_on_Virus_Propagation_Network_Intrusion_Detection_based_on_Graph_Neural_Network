```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='whitegrid')
```

```python
result_nslkdd_binary = pd.read_csv('./result/nslkdd/binary/total_result_nslkdd.csv')
result_kdd99_binary = pd.read_csv('./result/kdd99/binary/total_result_kdd99.csv')
result_nslkdd_five = pd.read_csv('./result/nslkdd/five/total_result_nslkdd.csv')
result_kdd99_five = pd.read_csv('./result/kdd99/five/total_result_kdd99.csv')
```

```python
fig = plt.figure(figsize=(20, 10))

ax1 = plt.subplot(2, 2, 1)
for idx, df in result_nslkdd_binary.groupby(['sample_rate']):
    plt.plot(df['feature_num'], df['acc'], '.-', label=idx)
plt.xlabel('Feature number\n(a)')
plt.ylabel('Accuracy')
plt.legend(title='Sample rate (%)', loc='right')
plt.title('Binary classification')

ax2 = plt.subplot(2, 2, 2)
floor = 15
for idx, df in result_nslkdd_binary.groupby(['sample_rate']):
    plt.plot(df['feature_num'].iloc[floor:], df['acc'].iloc[floor:], '.-', label=idx)
plt.xlabel('Feature number\n(b)')
plt.ylabel('Accuracy')
plt.xticks(range(floor, 42))
plt.legend(title='Sample rate (%)', loc='right')
plt.title('Binary classification')

ax3 = plt.subplot(2, 2, 3)
for idx, df in result_nslkdd_five.groupby(['sample_rate', 'feature_num']).mean().reset_index().groupby('sample_rate'):
    plt.plot(df['feature_num'], df['acc'], '.-', label=idx)
plt.xlabel('Feature number\n(c)')
plt.ylabel('Accuracy')
plt.legend(title='Sample rate (%)', loc='right')
plt.title('Five-class classification')

ax4 = plt.subplot(2, 2, 4)
floor = 20
for idx, df in result_nslkdd_five.groupby(['sample_rate', 'feature_num']).mean().reset_index().groupby('sample_rate'):
    plt.plot(df['feature_num'].iloc[floor:], df['acc'].iloc[floor:], '.-', label=idx)
plt.xlabel('Feature number\n(d)')
plt.ylabel('Accuracy')
plt.xticks(range(floor, 42))
plt.legend(title='Sample rate (%)', loc='right')
plt.title('Five-class classification')

plt.subplots_adjust(hspace=0.35)
```
